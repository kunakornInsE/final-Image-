```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
import urllib
```

```python
url="https://github.com/jbinteam/010723305/blob/main/images/test.jpg?raw=true"

url_response=urllib.request.urlopen(url)

img_array=np.array(bytearray(url_response.read()), dtype=np.uint8)
cv_img=cv2.imdecode(img_array,-1)

modifyer=np.array([0,0,1]) #BGR

output=cv_img*modifyer

images=np.hstack((cv_img,output))

print(cv_img.shape)
print(images.shape)

cv2_imshow(images)
```
